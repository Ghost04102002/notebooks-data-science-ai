```python
stores = stores.loc[stores.city!="Quito",:]	
stores = stores.loc[stores.city!="Guayaquil",:]
```

```python
dfene = dfene.merge(stores, on="store_nbr")
dffeb = dfene.merge(stores, on="store_nbr")
dfmar = dfene.merge(stores, on="store_nbr")
```

```python
dfene["date"] = pd.to_datetime(dfene.date)
dfene["day"] = dfene.date.dt.day_name()
table1 = pd.pivot_table(data=dfene, index="day", columns="state", values="unit_sales", aggfunc="sum").fillna(0)
```

```python
dffeb["date"] = pd.to_datetime(dffeb.date)
dffeb["day"] = dffeb.date.dt.day_name()
table2 = pd.pivot_table(data=dffeb, index="day", columns="state_x", values="unit_sales", aggfunc="sum").fillna(0)
```

```python
dfmar["date"] = pd.to_datetime(dfmar.date)
dfmar["day"] = dfmar.date.dt.day_name()
table3 = pd.pivot_table(data=dfmar, index="day", columns="state_x", values="unit_sales", aggfunc="sum").fillna(0)
```
