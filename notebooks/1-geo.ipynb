```python
import geopandas as gpd
import pandas as pd
```

```python
gdf = gpd.read_file("../input/parroquias.geojson")
```

```python
gdf
```

```python
gdf.plot()
```

```python
df = pd.read_csv("../output/data.csv")
```

```python
df
```

```python
gcasas = gpd.GeoDataFrame(df,geometry= gpd.points_from_xy(df["long"],df["lat"]))
```

```python
gcasas
```

```python
gcasas.plot()
```

```python
gcasas.crs = gdf.crs
```

```python
gjoin = gpd.sjoin(gcasas,gdf)
```

```python
gjoin
```

```python
gjoin.plot()
```

```python
gjoin.to_file("../output/gjoin.geojson", driver="GeoJSON")
```
