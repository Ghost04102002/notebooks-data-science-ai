```python
import pandas as pd
import numpy as np
```

```python
df =  pd.read_csv("../input/data_2019-1.csv")
```

```python
df.drop_duplicates(subset=["metroscuadradostotales","latlong"], keep="last", inplace = True)
```

```python
df.dropna(subset=["antigüedad","cuartos","cuartosdebaño","estacionamiento","latlong","metroscuadradostotales","precio","publicado","tipodevendedor","titulo"],inplace=True)
```

```python
df = df.loc[df.precio!=" ",:]
df.precio = df.precio.str.replace("$","").str.replace(".","",regex=False).astype("float64")
```

```python
df[["latitud","longitud"]] = df.latlong.str.split("place/",expand=True)[1].str.split(",",expand=True)
```

```python
df.dropna(subset=["latitud","longitud"],inplace=True)
```

```python
df.latitud = df.latitud.astype("float")
df.longitud = df.longitud.astype("float")
```

```python
df["preciopmt2"] = df["precio"] / df["metroscuadradostotales"]
```

```python
df.drop_duplicates(subset=["precio","metroscuadradostotales","latitud","longitud"], keep="last", inplace = True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.to_csv("../output/dataset_final_csv",index=False)
```
