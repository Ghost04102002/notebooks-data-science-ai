```python
def operaciones(row):
    if row["metroscuadradostotales"] != 0:
        v1 = row["precio"] / row["metroscuadradostotales"]
        v2 = row["precio"] // row["metroscuadradostotales"]
        return [v1,v2]
    else:
        return [0,0]
```

```python
df[["premt21","premt22"]] = df.apply(operaciones, axis = 1, result_type="expand")
```
