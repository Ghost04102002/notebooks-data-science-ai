```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
```

```python
gdf = gpd.read_file("../input/parroquias.geojson")
```

```python
df = pd.read_csv("../output/data.csv")
```

```python
df
```

```python
Q1 = df['precio'].quantile(0.25)
Q3 = df['precio'].quantile(0.75)
IQR = Q3 - Q1
```

```python
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
```

```python
df = df.loc[(df['precio'] >= limite_inferior)&(df['precio'] <= limite_superior),:]
```

```python
df
```

```python
df.drop_duplicates(subset=["metroscuadradostotales","lat","long"], keep="last", inplace = True)
df.dropna(subset=["metroscuadradostotales","precio","lat","long"],inplace=True)
df = df.loc[df["metroscuadradostotales"]>0,:]
df["preciopmt2"] = df["precio"] / df["metroscuadradostotales"]
```

```python
df
```

```python
gcasas = gpd.GeoDataFrame(df,geometry= gpd.points_from_xy(df["long"],df["lat"]))
```

```python
gcasas.crs = gdf.crs
gjoin = gpd.sjoin(gcasas,gdf)
grouped = gjoin.groupby('Name').agg({
    'precio': 'mean',  
    'preciopmt2': 'mean',
    'geometry': lambda x: x.unary_union.centroid  
}).reset_index()
gdfR = gpd.GeoDataFrame(grouped, geometry='geometry')
```

```python
gdfR
```

```python
gdfR.set_crs(epsg=4326, inplace=True)
```

```python
gdfR.to_file("data.geojson", driver='GeoJSON')
```
